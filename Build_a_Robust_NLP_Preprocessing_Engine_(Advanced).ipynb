**Task 1: Conceptual Understanding**

**1. What is the difference between "Love" and "love" in NLP?**

In Natural Language Processing, "Love" and "love" are typically treated as two distinct tokens due to case sensitivity. During the tokenization phase, a word is usually considered unique if its sequence of characters differs from another.
If the text is not converted to lowercase (case folding), a model may fail to recognize that the capitalized version (often appearing at the start of a sentence or as a proper noun) and the lowercase version refer to the same concept.

This can lead to an inflated vocabulary size and data sparsity, reducing the efficiency of the machine learning model.

**2. What happens if stopwords are not removed?**

If stopwords (common words like "the," "a," "is," "and") are not removed, they introduce noise and bias into the text data. These words carry very little specific semantic meaning but appear with high frequency, causing them to dominate frequency-based analysis and text representations.

Consequently, this can decrease the signal-to-noise ratio, making it harder for models to learn the relationships between the truly meaningful keywords and the document's content, especially in tasks like text classification or information retrieval.

**3. Provide two real-world scenarios where removing stopwords can be harmful.**

Sentiment Analysis and Negation Handling: Removing negation words, which are often considered stopwords (e.g., "not," "no," "don't"), can fundamentally change the meaning of a sentence, leading to incorrect sentiment classification. For example, "The product was not good" would lose the crucial word "not," potentially reversing the intended negative sentiment.

Machine Translation and Text Generation: In tasks requiring the preservation of grammatical structure, such as machine translation, removing stopwords destroys the syntactic integrity of the sentence. A sentence like "She walked to the store" might become "She walked store," which is grammatically incorrect and loses semantic clarity.


**4. Explain the difference between stemming and lemmatization.**

Stemming is a heuristic process that reduces a word to its root form (stem) by chopping off suffixes. It is generally faster and simpler to implement. However, the resulting stem is often not a true dictionary word and may be over- or under-stemmed (e.g., "caring" might be reduced to "car" or "universe" to "univers").

Lemmatization is a more sophisticated, dictionary-based process that uses morphological analysis (often requiring Part-of-Speech tagging) to reduce a word to its canonical base form (lemma). It is slower but guarantees that the resulting word is a valid word in the language's vocabulary (e.g., "better" is lemmatized to "good," and "running" to "run").

```python
import re
from collections import Counter
import numpy as np
```

```python
def preprocess_text(text):
    try:
        if not isinstance(text, str) or text.strip() == "":
            return [], ""

        # Remove URLs
        text = re.sub(r"http\S+|www\S+", "", text)

        # Remove emails
        text = re.sub(r"\S+@\S+", "", text)

        # Remove numbers
        text = re.sub(r"\d+", "", text)

        # Remove emojis (basic)
        text = re.sub(r"[^\w\s]", "", text)

        # Lowercase
        text = text.lower()

        # Handle repeated characters (soooo → soo)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        tokens = text.split()

        # Remove short tokens (≤2) except "no", "not"
        tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

        cleaned_sentence = " ".join(tokens)

        return tokens, cleaned_sentence

    except:
        return [], ""
```

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
all_tokens = []

for s in sentences:
    tokens, clean = preprocess_text(s)
    all_tokens.extend(tokens)

    print("Original:", s)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-" * 50)
```

```python
for s in sentences:
    tokens, _ = preprocess_text(s)

    if len(tokens) == 0:
        continue

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = np.mean([len(t) for t in tokens])

    print(f"Sentence: {s}")
    print(f"Total Tokens: {total}")
    print(f"Unique Tokens: {unique}")
    print(f"Avg Token Length: {round(avg_len,2)}")
    print("-" * 50)
```

### Analysis

1. Sentence with most noise:
"Win $$$ now!!! Limited offer!!!" because it contains symbols, repeated punctuation, and promotional noise.

2. Sentence with most meaningful tokens:
"I am not happy with this" because it retains clear sentiment-related words even after preprocessing.

```python
counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(sentences)
print(result)
```
