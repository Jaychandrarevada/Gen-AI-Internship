```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Identify the highest performance score
highest_score = max(employees.values())

# Find all employees with the highest score
top_performers = [
    name for name, score in employees.items() if score == highest_score
]

# Format the output
output_message = (
    f"Top Performers Eligible for Bonus: {', '.join(top_performers)} "
    f"(Score: {highest_score})"
)

# Display the result
print(output_message)
```

```python
import collections
import string

search_query = "Buy mobile phone buy phone online"

# Convert to lowercase
search_query = search_query.lower()

# Remove punctuation
search_query = search_query.translate(str.maketrans('', '', string.punctuation))

# Split into keywords
keywords = search_query.split()

# Count the frequency of each keyword
keyword_counts = collections.Counter(keywords)

# Display only keywords searched more than once
result = {word: count for word, count in keyword_counts.items() if count > 1}

print(result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Initialize a list to store valid readings
valid_readings = []

# Iterate through the sensor readings with their indices
for hour_index, reading_value in enumerate(sensor_readings):
    # Check if the reading is an even number
    if reading_value % 2 == 0:
        # Store as (hour_index, reading_value) pair
        valid_readings.append((hour_index, reading_value))

# Display the result
print(f"Valid Sensor Readings (Hour, Value):\n{valid_readings}")
```

```python
import collections

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Extract domains from email addresses
domains = [email.split('@')[1] for email in emails]

# Count the frequency of each domain
domain_counts = collections.Counter(domains)

# Calculate the total number of emails
total_emails = len(emails)

# Calculate and display the percentage usage of each domain
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate the average daily sales
average_sales = sum(sales) / len(sales)

print(f"Average Daily Sales: {average_sales:.2f}")
print("Days with Sales Spikes:")

# Detect days where sales are more than 30% above average
for day_index, sale_value in enumerate(sales):
    # Day numbers are 1-indexed for the output
    day_number = day_index + 1
    if sale_value > average_sales * 1.30:
        print(f"Day {day_number}: {sale_value}")
```

```python
import collections

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count the frequency of each user ID
user_id_counts = collections.Counter(user_ids)

# Identify and display duplicate user IDs and their counts
print("Duplicate User IDs:")
for user_id, count in user_id_counts.items():
    if count > 1:
        print(f"{user_id} \u2192 {count} times")
```
