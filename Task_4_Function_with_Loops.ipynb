```python
def validate_recharge_request():
    valid_plans = {199, 299, 399, 599}

    while True:
        try:
            recharge_amount_str = input("Enter recharge amount (minimum ₹50): ₹")
            recharge_amount = int(recharge_amount_str)

            if recharge_amount < 50:
                print("Recharge amount must be at least ₹50. Please try again.")
                continue

            plan_str = input(f"Enter plan (valid plans: {', '.join(map(str, valid_plans))}): ")
            plan = int(plan_str)

            if plan not in valid_plans:
                print(f"Invalid plan. Valid plans are: {', '.join(map(str, valid_plans))}. Please try again.")
                continue

            print(f"Recharge request valid! Amount: ₹{recharge_amount}, Plan: ₹{plan}")
            break

        except ValueError:
            print("Invalid input. Please enter a number for amount and plan.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

# Call the function to start the validation process
validate_recharge_request()
```

```python
def check_inventory(products_stock):
    reorder_threshold = 15

    print("\n--- Inventory Status ---")
    for product, stock_quantity in products_stock.items():
        if stock_quantity < reorder_threshold:
            print(f"Reorder Alert: {product} (Stock: {stock_quantity})")
        else:
            print(f"Stock OK: {product} (Stock: {stock_quantity})")
    print("------------------------")

# Sample product inventory
products = {
    "Laptop": 25,
    "Mouse": 10,
    "Keyboard": 30,
    "Monitor": 12,
    "Webcam": 50
}

# Call the function to check inventory
check_inventory(products)
```

```python
def process_student_results(marks):
    if not marks:
        print("No marks provided.")
        return

    total_marks = sum(marks)
    average_marks = total_marks / len(marks)

    print(f"Student Marks: {marks}")
    print(f"Average Marks: {average_marks:.2f}")

    if average_marks >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")

# Sample usage:
student1_marks = [60, 75, 45, 80, 55]
print("\n--- Student 1 Results ---")
process_student_results(student1_marks)

student2_marks = [30, 40, 25, 35, 48]
print("\n--- Student 2 Results ---")
process_student_results(student2_marks)

student3_marks = [50, 50, 50]
print("\n--- Student 3 Results ---")
process_student_results(student3_marks)

student4_marks = []
print("\n--- Student 4 Results ---")
process_student_results(student4_marks)
```

```python
def calculate_cab_fare():
    base_fare = 50
    per_km_rate = 12
    peak_hour_surcharge_percentage = 0.25

    while True:
        try:
            distance_str = input("Enter distance traveled in km: ")
            distance = float(distance_str)

            if distance < 0:
                print("Distance cannot be negative. Please try again.")
                continue

            is_peak_hour_str = input("Is it peak hour? (yes/no): ").lower()
            if is_peak_hour_str not in ['yes', 'no']:
                print("Invalid input for peak hour. Please enter 'yes' or 'no'.")
                continue

            fare = base_fare + (distance * per_km_rate)

            if is_peak_hour_str == 'yes':
                fare += fare * peak_hour_surcharge_percentage

            print(f"Estimated Cab Fare: ₹{fare:.2f}")
            break

        except ValueError:
            print("Invalid input. Please enter a valid number for distance.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

# Call the function to start the fare estimation process
calculate_cab_fare()
```

```python
def check_attendance_eligibility(attendance_list):
    if not attendance_list:
        print("No attendance records provided.")
        return "Not Eligible (No records)"

    total_days = len(attendance_list)
    present_days = attendance_list.count('P')

    attendance_percentage = (present_days / total_days) * 100

    print(f"Attendance Records: {attendance_list}")
    print(f"Present Days: {present_days}/{total_days}")
    print(f"Attendance Percentage: {attendance_percentage:.2f}%")

    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Sample usage:
student1_attendance = ['P', 'P', 'A', 'P', 'P', 'P', 'A', 'P', 'P', 'P'] # 80% (8/10)
print("\n--- Student 1 Attendance ---")
result1 = check_attendance_eligibility(student1_attendance)
print(f"Result: {result1}")

student2_attendance = ['P', 'P', 'A', 'P', 'A', 'A', 'P', 'P'] # 62.5% (5/8)
print("\n--- Student 2 Attendance ---")
result2 = check_attendance_eligibility(student2_attendance)
print(f"Result: {result2}")

student3_attendance = ['P', 'P', 'P', 'P'] # 100% (4/4)
print("\n--- Student 3 Attendance ---")
result3 = check_attendance_eligibility(student3_attendance)
print(f"Result: {result3}")

student4_attendance = []
print("\n--- Student 4 Attendance ---")
result4 = check_attendance_eligibility(student4_attendance)
print(f"Result: {result4}")
```

```python
def check_password_strength(password):
    min_length = 8
    has_digit = False
    has_special_char = False
    special_characters = "@#$"

    # Check minimum length
    if len(password) < min_length:
        print(f"Password is too short. Minimum length is {min_length}.")
        return "Weak"

    # Check for digit and special character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special_char = True

    if has_digit and has_special_char:
        print("Password strength: Strong")
        return "Strong"
    elif has_digit or has_special_char:
        print("Password strength: Moderate (consider adding more complexity)")
        return "Moderate"
    else:
        print("Password strength: Weak (add digits and special characters)")
        return "Weak"

# Sample usage:
print("\n--- Password Checks ---")
check_password_strength("MyPass123@") # Strong
check_password_strength("mypasword") # Weak
check_password_strength("mypass123") # Moderate (missing special char)
check_password_strength("MyPass@") # Weak (too short)
check_password_strength("MyPass@1") # Strong
```
