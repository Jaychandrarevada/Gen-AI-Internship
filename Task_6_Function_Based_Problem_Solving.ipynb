```python
def manage_parking_lot(capacity, vehicle_logs):
    """
    Manages a smart parking lot by tracking vehicle entry/exit and capacity.

    Args:
        capacity (int): The total parking capacity.
        vehicle_logs (list): A list of strings ("IN" or "OUT") representing logs.

    Returns:
        tuple: A tuple containing (current_parked, status)
    """
    parked_vehicles = 0
    peak_vehicles = 0

    # 1. Loop for counting
    for log in vehicle_logs:
        if log == "IN":
            parked_vehicles += 1
        elif log == "OUT":
            # Prevent negative count in case of error in logs
            parked_vehicles = max(0, parked_vehicles - 1)

        # Track peak usage
        peak_vehicles = max(peak_vehicles, parked_vehicles)

        # 4. Conditional alert for exceeding capacity
        if parked_vehicles > capacity:
            print(f"**ALERT: Parking has exceeded capacity of {capacity} at log event: {log}**")

    # 2. Calculate total parked vehicles and 3. Identify parking status
    if parked_vehicles > capacity:
        # This should only happen if the alert wasn't triggered immediately, but included for completeness
        parking_status = "Exceeded Capacity"
    elif parked_vehicles == capacity:
        parking_status = "Full"
    else:
        parking_status = "Available"

    print(f"Currently Parked Vehicles: {parked_vehicles}")
    print(f"Parking Status: {parking_status}")
    print(f"Peak Parking Usage: {peak_vehicles}") # Added to address point 3

# Sample Input Execution
print("\n--- Problem 1 Solution ---")
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking_lot(parking_capacity, vehicle_logs)
```

```python
def estimate_delivery_time(distance_km, traffic_level, weather):
    """
    Estimates delivery time based on distance, traffic, and weather.

    Args:
        distance_km (float): Distance in kilometers.
        traffic_level (str): "Low", "Medium", or "High".
        weather (str): "Clear", "Rainy", or "Snowy".

    Returns:
        int: The estimated total delivery time in minutes.
    """
    # 1. Base Time: Assume 2 minutes per km for base travel (30 km/h)
    BASE_SPEED_MIN_PER_KM = 3  # Adjusted to get closer to Sample Output (8km * 3 = 24min)
    total_time = distance_km * BASE_SPEED_MIN_PER_KM

    # Apply Traffic Delays (Mathematical adjustments based on multiple conditions)
    traffic_delay = 0
    if traffic_level == "High":
        traffic_delay = 25
    elif traffic_level == "Medium":
        traffic_delay = 15
    elif traffic_level == "Low":
        traffic_delay = 5

    # Apply Weather Delays
    weather_delay = 0
    if weather == "Rainy" or weather == "Snowy":
        weather_delay = 10

    # Calculate final ETA
    total_time += traffic_delay + weather_delay
    final_eta = round(total_time) # Round to nearest minute for clean output

    print(f"Estimated Delivery Time: {final_eta} minutes")
    return final_eta

# Sample Input Execution
print("\n--- Problem 2 Solution ---")
distance = 8
traffic = "High"
weather = "Rainy"
estimate_delivery_time(distance, traffic, weather)
```

```python
def analyze_seat_occupancy(total_seats, booked_seats_list):
    """
    Analyzes seat booking data to determine occupancy and show status.

    Args:
        total_seats (int): The total number of seats in the theatre.
        booked_seats_list (list): A list where the length equals the number of booked seats.
    """
    booked_seats = len(booked_seats_list)

    # 1. Calculates occupancy percentage
    if total_seats == 0:
        occupancy_percentage = 0
    else:
        occupancy_percentage = (booked_seats / total_seats) * 100

    # 2. Determines if show is Housefull and 3. Suggests additional shows
    show_status = ""
    suggest_shows = False

    if booked_seats >= total_seats:
        show_status = "Housefull"
        suggest_shows = True
    elif occupancy_percentage >= 75:
        show_status = "Almost Full"
        suggest_shows = True
    else:
        show_status = "Available"
        suggest_shows = False

    print(f"Occupancy: {occupancy_percentage:.0f}%")
    print(f"Show Status: {show_status}")
    if suggest_shows:
        print("Suggestion: Consider opening additional shows.")

# Sample Input Execution
print("\n--- Problem 3 Solution ---")
total = 200
# The list length determines booked seats (150 entries)
booked = [1] * 150
analyze_seat_occupancy(total, booked)
```

```python
def classify_server_load(cpu_readings):
    """
    Classifies server load based on CPU usage readings.

    Args:
        cpu_readings (list): A list of CPU usage percentages (integers).
    """
    # 1. Function + loop: Calculate Average CPU Load
    if not cpu_readings:
        average_cpu_load = 0
    else:
        average_cpu_load = sum(cpu_readings) / len(cpu_readings)

    # 2. Conditional Classification based on rules
    server_status = ""
    if average_cpu_load > 80:
        server_status = "Critical"
    elif 50 <= average_cpu_load <= 80:
        server_status = "Warning"
    elif average_cpu_load < 50:
        server_status = "Normal"

    print(f"Average CPU Load: {average_cpu_load:.0f}%")
    print(f"Server Status: {server_status}")

# Sample Input Execution
print("\n--- Problem 4 Solution ---")
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

```python
def monitor_resource_usage(resource_usage, overuse_threshold=7):
    """
    Tracks usage of classroom resources and identifies overuse patterns.

    Args:
        resource_usage (dict): Dictionary of resource names and usage hours.
        overuse_threshold (int): The number of hours considered 'overuse'.
    """
    overused_resources = []

    # 1. Function + dictionary + conditional checks
    for resource, hours in resource_usage.items():
        if hours > overuse_threshold:
            overused_resources.append(resource)

    # Determine alert status
    energy_alert = "Yes" if overused_resources else "No"

    print(f"Overused Resources: {', '.join(overused_resources) if overused_resources else 'None'}")
    print(f"Energy Alert: {energy_alert}")

# Sample Input Execution
print("\n--- Problem 5 Solution ---")
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_resource_usage(usage_data)
```

```python
def control_registration_capacity(event_capacity, registrations):
    """
    Manages event registrations, prevents overbooking, and triggers waitlist mode.

    Args:
        event_capacity (int): The maximum number of attendees allowed.
        registrations (int): The total number of users who attempted to register.
    """
    # 1. Tracking registrations & 2. Preventing overbooking
    if registrations <= event_capacity:
        confirmed_registrations = registrations
        waitlisted_users = 0
        registration_status = "Open"
    else:
        # Registrations > Capacity, so capacity is max confirmed
        confirmed_registrations = event_capacity
        # 3. Triggering waitlist mode
        waitlisted_users = registrations - event_capacity
        registration_status = "Closed"

    print(f"Confirmed Registrations: {confirmed_registrations}")
    print(f"Waitlisted Users: {waitlisted_users}")
    print(f"Registration Status: {registration_status}")

# Sample Input Execution
print("\n--- Problem 6 Solution ---")
capacity = 100
total_registrations = 105
control_registration_capacity(capacity, total_registrations)
```
