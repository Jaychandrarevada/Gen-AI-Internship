### 1. User Login Check

**Problem Statement:** Given a username and password, check whether login is successful.

**Requirements:**
*   Print "Login Successful" if both username and password match
*   Otherwise print "Invalid Credentials"

```python
# Define the correct username and password
correct_username = "admin"
correct_password = "1234"

# User input (for demonstration, using hardcoded values)
# username = input("Enter username: ")
# password = input("Enter password: ")

username = "admin"
password = "1234"

# Check if both username and password match
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")

# Test with invalid credentials
username_invalid = "admin"
password_invalid = "wrong_pass"

if username_invalid == correct_username and password_invalid == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

### 2. Pass / Fail Analyzer

**Problem Statement:** Analyze student results.

**Requirements:**
*   A student passes if marks â‰¥ 50
*   Count the total number of pass students
*   Count the total number of fail students
*   Print the final result clearly

```python
# Given student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Iterate through the marks to check each student's result
for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print the final result
print(f"Total students: {len(marks)}")
print(f"Number of students who passed: {pass_count}")
print(f"Number of students who failed: {fail_count}")
```

### 3. Simple Data Cleaner

**Problem Statement:** Clean and standardize user names.

**Requirements:**
*   Remove extra spaces
*   Convert all names to lowercase
*   Print the cleaned list

```python
# Given list of names
names = [" Alice ", "bob", " CHARLIE "]

# Create an empty list to store cleaned names
cleaned_names = []

# Iterate through each name, clean it, and add to the new list
for name in names:
    # Remove extra spaces using .strip() and convert to lowercase using .lower()
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print the cleaned list
print("Original names:", names)
print("Cleaned names:", cleaned_names)
```

### 4. Message Length Analyzer

**Problem Statement:** Analyze message sizes.

**Requirements:**
*   Print the length of each message
*   Flag messages longer than 10 characters

```python
# Given list of messages
messages = ["Hi", "Welcome to the platform", "OK", "Hello World!"]

# Iterate through each message
for message in messages:
    message_length = len(message)
    print(f"Message: '{message}' | Length: {message_length}", end='')

    # Flag messages longer than 10 characters
    if message_length > 10:
        print(" (Long message)")
    else:
        print()
```

### 5. Error Message Detector

**Problem Statement:** Detect error messages from system logs.

**Requirements:**
*   Count the number of "ERROR" entries
*   Print the total error count

```python
# Given list of log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize error counter
error_count = 0

# Iterate through the log entries to count "ERROR" messages
for entry in logs:
    if entry == "ERROR":
        error_count += 1

# Print the total error count
print(f"Total number of 'ERROR' entries: {error_count}")
```
