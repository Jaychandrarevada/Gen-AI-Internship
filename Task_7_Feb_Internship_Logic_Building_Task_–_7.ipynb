```python
class SmartLight:
    """
    A class to control a smart light (ON/OFF) and display its status.
    """
    def __init__(self, name):
        self.name = name
        self.status = "OFF"  # Default status

    def turn_on(self):
        self.status = "ON"
        self.display_status()

    def turn_off(self):
        self.status = "OFF"
        self.display_status()

    def display_status(self):
        print(f"{self.name} is {self.status}")

# Sample Usage
bedroom_light = SmartLight("Bedroom Light")
bedroom_light.turn_on()
```

```python
class Employee:
    """
    A class that stores employee details and displays an ID card.
    """
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.employee_id}")
        print(f"Department: {self.department}")

# Sample Usage
employee1 = Employee("Rahul", "EMP102", "IT")
employee1.display_id_card()
```

```python
class MobileContact:
    """
    A class to store contact name and phone number and display information.
    """
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number
        self.save_contact()

    def save_contact(self):
        print("Contact Saved")
        self.display_contact()

    def display_contact(self):
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

contact1 = MobileContact("Anita", "9876543210")
```

```python
class Product:
    """
    A class that stores product name and price and prints a formatted price tag.
    """
    def __init__(self, name, price):
        self.name = name
        # Convert price to integer for clean formatting
        self.price = int(price)

    def generate_price_tag(self):
        # Format the price with the Indian Rupee symbol (₹) and a comma separator
        formatted_price = f"₹{self.price:,}"
        print(f"Product: {self.name}")
        print(f"Price: {formatted_price}")


product1 = Product("Headphones", 2499)
product1.generate_price_tag()
```

```python
class Movie:
    """
    A class that stores movie name and rating and displays movie details.
    """
    def __init__(self, name, rating):
        self.name = name
        # Rating is assumed to be out of 5 based on the expected output
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        # Format the rating to display "/ 5"
        print(f"Rating: {self.rating} / 5")

movie1 = Movie("Inception", 4.8)
movie1.display_details()
```

```python
class DeliveryManager:
    """
    A class that stores customer name and delivery address and prints delivery details.
    """
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

delivery1 = DeliveryManager("Jaychandra", "Hyderabad")
delivery1.print_delivery_details()
```
